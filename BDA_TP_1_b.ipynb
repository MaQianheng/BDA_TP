```python
import nltk
from pyspark import SparkConf, SparkContext
from pyspark.sql import SQLContext, SparkSession
from nltk.corpus import stopwords
import string
```

```python
def word_tokenize(line):
	if isinstance(line, str):
		return nltk.word_tokenize(line)
	else :
		return nltk.word_tokenize(str(line))
```

```python
def pos_tag(word):
#     nltk.download('averaged_perceptron_tagger')
    return nltk.pos_tag([word])
```

```python
def handle_csv(path):
#     'hdfs://cluster001:9000/user/input/test.csv'
	data_frame_text = spark_session.read.csv(path)

	# keep the useful columns
	data_frame_text = data_frame_text.select('_c2','_c3','_c4')
	# convert to rdd
	rdd_text = data_frame_text.rdd.map(lambda x: [str(x[0]), x[1], x[2]])

	# remove the first line
	header_text = rdd_text.first()
	rdd_text_noheader = rdd_text.filter(lambda row: row != header_text)
	return rdd_text_noheader
```

```python
def containsAny(seq,aset):
    for c in seq:
         if c in aset:
                return True
    return False
```

```python
count = 0
final_data = ''
for file_name in files_name:
    inputfile_dir = 'hdfs://cluster001:9000/user/input/' + file_name
    tmp_rdd = handle_csv(inputfile_dir)
    if count==0:
        final_data = tmp_rdd
    else:
        final_data = final_data.union(tmp_rdd)
    count += 1
```

```python
final_data.count()
```

```python
stop_words=set(stopwords.words('english'))
stop_words.add('http')
stop_words.add('https')
stop_words.add('null')
list_punct=list(string.punctuation) + ['’','“','”','‘','—']
key_words = ['Thai', 'Thailand', 'thai', 'thailand']
```

```python
def contains_any(arr):
	for element in arr:
		if isinstance(element, str):
			for key_word in key_words:
				if key_word in element:
					return True
		else :
			for key_word in key_words:
				if key_word in str(element):
					return True
	return False
```

```python
classified_text_rdd = final_data \
.map(lambda line :line) \
.filter(contains_any) \
.flatMap(lambda row: row) \
.flatMap(word_tokenize) \
.map(lambda word: word.lower()) \
.filter(lambda word : word not in stop_words and word != '') \
.filter(lambda word : word not in list_punct) \
.map(pos_tag) \
.filter(lambda word : word[0][1][0] == 'N') \
.map(lambda word: word[0][0]) \
.map(lambda word: (word, 1)) \
.reduceByKey(lambda a, b: a+b) \
.sortBy(lambda a: a[1], ascending = False)
```

```python
# print(classified_text_rdd.collect())
classified_text_rdd \
.saveAsTextFile('hdfs://cluster001:9000/user/output/BDA_TP_1_b')
```
